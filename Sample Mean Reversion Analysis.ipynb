# Sample Mean Reversion Analysis
This notebook uses the pipeline from the Sample Mean Reversion example to find the most prfitable stocks in a year

run_pipeline form quantopian.research allows to execute a pipeline a number of days in a notebook

```python
from quantopian.research import run_pipeline
```

```python
from quantopian.pipeline import Pipeline
from quantopian.pipeline.factors import Returns
from quantopian.pipeline.filters import QTradableStocksUS
```

```python
RETURNS_LOOKBACK_DAYS = 5
```

```python
def make_pipeline(): # the variable context from the algorithm pipeline is removed
    
    universe = QTradableStocksUS()
    
    recent_returns = Returns(
        window_length=RETURNS_LOOKBACK_DAYS,
        mask=universe
    )
    
    recent_returns_zscore = recent_returns.zscore()

    
    low_returns = recent_returns_zscore.percentile_between(0,10)
    high_returns = recent_returns_zscore.percentile_between(90,100)

    securities_to_trade = (low_returns | high_returns)

    pipe = Pipeline(
        columns={
            'recent_returns_zscore': recent_returns_zscore
        },
        screen=securities_to_trade
    )

    return pipe
```

```python
my_pipe = make_pipeline()
my_pipe
```
